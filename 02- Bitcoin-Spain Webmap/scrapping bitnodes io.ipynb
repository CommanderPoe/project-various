```python
# Scraping bitnodes
from bs4 import BeautifulSoup
import requests
import regex as re
```

```python
# Testing first page
url = 'https://bitnodes.io/nodes/?q=Spain'
```

```python
# checking response of the website
response = requests.get(url)
response.status_code
```

```python
# parsing the website (only single page) 
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# selecting table with info and printing results
table = soup.select('body > div.container > div:nth-child(10) > div > table > tr')

for i in table:
    print(i.get_text())
```

```python
# Saving text in an empty list
text = [] 

for i in table: 
    text.append(i.get_text()) 
```

<br/>

```python
# Headers
headers = text[0].split('\n')[2:-1] 
headers = [item.lower().replace(' ','_') for item in headers]
```

```python
headers
```

<br/>

```python
# Ip addresses
address = []


for n in range(1,len(text)):
    #if end_ip in text[n][2:17]
    address.append(re.findall('[\d[.]+', text[n])[0])
        
```

```python
address
len(address)
```

<br/>

```python
# User Agent
user_agent = []


for n in range(1,len(text)):
    user_agent.append(re.findall('Sa.+\d',text[n])[0])
```

```python
user_agent
len(user_agent)
```

<br/>

```python
# Height
height = []


for n in range(1,len(text)):
    height.append(re.findall('\d{5}', text[1])[0])
```

```python
height
len(height)
```

```python
# Location
location = []

for n in range(1,len(text)):
    location.append(re.findall('\w+, Sp.+', text[n])[1])    
```

```python
new_loc = []

for n in range(len(location)):
    new_loc.append(location[n].replace(' ','').replace('/',',').split(','))
```
